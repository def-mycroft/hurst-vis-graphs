```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# create data from lacasa paper
y = [87,49,36,83,]*5
y = np.array([x/100 for x in y]) 
n = len(y)
x = np.array(range(n)) 
```

```python
# plot data from lacasa paper
FIGSIZE = (12,4)
fig = plt.figure(figsize=FIGSIZE)
ax = plt.subplot(111)
ax.bar(x,y)
ax.xaxis.set(ticks=x)
ax.grid()
```
